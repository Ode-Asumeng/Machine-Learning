#### Declaring your IV and DV

```python
x = square.iloc[:, :1]
y = square.iloc[:, 1:2]
```

### Data Preprocessing

#### Interpolate .interpolate() looks at the pattern of the dataset and estimates the missing values based on the pattern

```python
X = x.interpolate()
X
```

#### To fix the duplicate values

#### We use the mask to group our data while maintaining the original structure of our data to be able to make changes in their original structure

```python
#we used keep=False to make the duplicates empty and not replacing it with any duplicate value.
xs=X.mask(X.duplicated(keep=False))
x=xs.interpolate()
x
```

### Another approach to preprocessing our data by combing the steps used in the approach above

```python
nums=square['x'].mask(square['x'].duplicated(keep=False))
#nums.interpolate()
square.interpolate()
```

### Splitting our data in to test and train sets

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 0)
```

### Training the algorithim using Linear Regression

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

#### Calling out our predictions

```python
y_pred = regressor.predict(x_test)
pd.DataFrame(y_pred)
```
